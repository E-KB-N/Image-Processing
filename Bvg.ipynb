```python
# import libraries
from skimage import io
from skimage import color
from skimage import data
from skimage.color import rgb2gray
from pylab import*
```

```python
# Read image
img = io.imread('bvg.jpg')
```

```python
# Convert to HSV
img_hsv = color.rgb2hsv(img)
```

```python
# Convert to RGB
img_rgb = color.hsv2rgb(img_hsv)
```

```python
# Convert to Grayscale
img_gs = color.rgb2gray(img)
```

```python
#Show original image
figure(0)
io.imshow(img)
```

```python
# Show conversion to HSV
io.imshow(img_hsv)
```

```python
# Display rgb version of image
io.imshow(img_rgb)
```

```python
# Display grayscale version of image
io.imshow(img_gs)
```
